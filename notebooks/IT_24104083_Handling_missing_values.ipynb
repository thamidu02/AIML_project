## Handling Missing Data

```python
# Identifying the missing data
data_df.isnull().sum()
```

```python
# Number of duplicated rows in the dataset
data_df.duplicated().sum()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data_df.isnull(),cbar=False)
plt.title('Missing Values Heatmap')
plt.show()


missing_counts=data_df.isnull().sum().sort_values(ascending=False)
print(missing_counts[missing_counts>0])
```

```python
missing_counts = data_df.isnull().sum()
missing_cols = missing_counts[missing_counts > 0].sort_values(ascending=False)


if len(missing_cols) == 0:
    print("No missing values in the dataset!")
else:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing_cols.index, y=missing_cols.values, palette='viridis')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Number of Missing Values')
    plt.title('Missing Values Count by Feature')
    plt.tight_layout()
    plt.show()
```

```python
missing_cols=['alcohol_consumption','caffeine_intake','exercise_type','insulin',
              'heart_rate','gene_marker_flag','income','daily_steps','blood_pressure']

print(data_df[missing_cols].dtypes)
```

```python
missing_cols=['blood_pressure','heart_rate','insulin','daily_steps','income','gene_marker_flag']
missing_cat_cols=data_df[missing_cols].select_dtypes(include=['object']).columns.tolist()
missing_num_cols=data_df[missing_cols].select_dtypes(include=['int64','float64']).columns.tolist()
print(missing_cat_cols)
print(missing_num_cols)
```

### Using model based imputation to fill the missing valuesa

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from xgboost import XGBRegressor

xgb = XGBRegressor(
    random_state=42
)
```

```python
# predicting the values for missing numeric data
imputer=IterativeImputer(estimator=xgb,random_state=42)
data_df[missing_num_cols]=imputer.fit_transform(data_df[missing_num_cols])
```

```python
missing_counts = data_df.isnull().sum()
missing_cols = missing_counts[missing_counts > 0].sort_values(ascending=False)


if len(missing_cols) == 0:
    print("No missing values in the dataset!")
else:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing_cols.index, y=missing_cols.values, palette='viridis')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Number of Missing Values')
    plt.title('Missing Values Count by Feature')
    plt.tight_layout()
    plt.show()
```
